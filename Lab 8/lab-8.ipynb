```python
import cv2
import os
from deepface import DeepFace
import numpy as np
import time
```

```python
KNOWN_FACES_DIR = 'Face testing'
```

```python
known_faces = []
known_names = []

for person_name in os.listdir(KNOWN_FACES_DIR):
    person_folder = os.path.join(KNOWN_FACES_DIR, person_name)
    
    if not os.path.isdir(person_folder):
        continue
    
    for image_name in os.listdir(person_folder):
        image_path = os.path.join(person_folder, image_name)
        face = DeepFace.represent(img_path=image_path, model_name="VGG-Face")
        known_faces.append(face)
        known_names.append(person_name)
```

```python
last_recognition_time = 0 
```

```python

video_capture = cv2.VideoCapture(0)

while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    current_time = time.time()

    #Run the model every 2 seconds and not every frame so my pc can survive
    if current_time - last_recognition_time >= 2:
        last_recognition_time = current_time

        # Perform face recognition
        result = DeepFace.find(frame, db_path=KNOWN_FACES_DIR, model_name="VGG-Face", enforce_detection=False)

        unique_identities = set()
        for df in result:
            if not df.empty:
                identities = df['identity'].tolist()
                unique_identities.update(identities)  

                best_match = df.loc[df['distance'].idxmin()]
                recognized_name = os.path.basename(os.path.dirname(best_match['identity']))
            else:
                print("No matches found.") #Does not work and i got too much on my hand so ill handle that in the assignment

        print("Face Match:", list(unique_identities))  # Convert set to list for display

    cv2.imshow('Face Recognition', cv2.resize(frame, (640, 480)))

    if cv2.waitKey(1) & 0xFF == ord('\r'): 
        break

# Release resources
video_capture.release()
cv2.destroyAllWindows()
```
