```python
def gradient_descent(x, y, m_current=0, b_current=0, epochs=1000, learning_rate=0.01):
    N = float(len(y))
    # Initialize lists to store the progression of parameter values and cost
    m_progression = []
    b_progression = []
    cost_progression = []
    
    for _ in range(epochs):
        y_current = (m_current * x) + b_current
        cost = sum([data**2 for data in (y-y_current)]) / N
        m_gradient = -(2/N) * sum(x * (y - y_current))
        b_gradient = -(2/N) * sum(y - y_current)
        
        m_current -= (learning_rate * m_gradient)
        b_current -= (learning_rate * b_gradient)
        
        # Store the current values
        m_progression.append(m_current)
        b_progression.append(b_current)
        cost_progression.append(cost)
    
    return m_current, b_current, cost, m_progression, b_progression, cost_progression

# Run gradient descent
m, b, cost, m_progression, b_progression, cost_progression = gradient_descent(x, y)
```

```python
import matplotlib.pyplot as plt

# Plotting the progression of Cost
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.plot(cost_progression, color='blue')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Cost Progression')

# Plotting the progression of Slope (m)
plt.subplot(1, 3, 2)
plt.plot(m_progression, color='red')
plt.xlabel('Iteration')
plt.ylabel('Slope (m)')
plt.title('Slope Progression')

# Plotting the progression of Intercept (b)
plt.subplot(1, 3, 3)
plt.plot(b_progression, color='green')
plt.xlabel('Iteration')
plt.ylabel('Intercept (b)')
plt.title('Intercept Progression')

plt.tight_layout()
plt.show()
```

```python
print("After Gradient Descent: ")
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
print(f"Cost: {cost}")
```
